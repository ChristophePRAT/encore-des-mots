```python
import pandas as pd
import os
from pathlib import Path    
from unidecode import unidecode
```

```python
# Set the directory path ('.' is current directory)
directory = Path('.')
```

```python
txt_files = list(directory.glob("*.txt"))
```

```python
txt_files
```

```python
for t in txt_files:
    # text = t.read_text()
    print(t.name[:-4])
```

```python
texts = {}

for t in txt_files:
    texts[t.name[:-4]] = t.read_text().split("\n")
```

```python
texts
```

```python
for key in texts:
    for i in range(len(texts[key])):
        trans = texts[key][i].split(",")[0].split("(")[0]
        texts[key][i] = trans
```

```python
texts["la"] = texts["la"][:-1] # drop last so that every body has 208 values
```

```python
for key in texts:
    print(f"{key} has {len(texts[key])} values")
```

```python
df = pd.DataFrame.from_dict(texts)
```

```python
df.head()
```

```python
latin = df.drop(["ru", "uk", "sr", "mk", "bg", "be", "cu", "cs"], axis=1) #remove non-latin alphabets
```
